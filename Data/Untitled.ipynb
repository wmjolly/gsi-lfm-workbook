```python
d = 'G:\My Drive\Work\Live Fuel Research\Data\LengthPIENClean.txt'
dat = pd.read_csv(d)
```

```python
dist = []
for idx,row in dat.iterrows():
    for i in range(0,int(row['length'])):
        dist.append(row['MidpointDiameter'])
        
    print(idx)
```

```python
import pandas
import matplotlib.pyplot as plt

vangel = dist
N = len(vangel)
plt.hist(vangel, density=False, alpha=0.5)
plt.title("Vangel material data (n={})".format(N))
plt.xlabel("Specimen strength")
```

```python
dist = np.array(dist)
shape, loc, scale = weibull_min.fit(dist, floc=0)
x = numpy.linspace(dist.min(), dist.max(), 100)
plt.plot(x, weibull_min(shape, loc, scale).pdf(x))
plt.title("Weibull fit fraction of total fuel length")
plt.xlabel("Midpoint Diameter (mm)")
plt.grid()
plt.savefig("../Figures/DensityVsLFM_PSEMEN_BlueMT_2021.png")
```
